```python
def fit(fit_data, plot_data, quadratic=False):
    def fit_func_quadratic(x, a, b, c, d):
        tblsizes = [i[0] for i in x]
#         cor_tbl_sizes = np.multiply(np.square(tblsizes),0.4459855440)
#         cor_tbl_sizes = np.add(np.multiply(tblsizes, 0.4459855440), tblsizes)
#         cor_tbl_sizes = np.multiply([math.log(i*100000, 2) for i in tblsizes], np.multiply(tblsizes, 0.4459855440))
        hash_tbl_sizes = np.multiply(0.4459855440, tblsizes)
        instances = [i[1] for i in x]
        return np.multiply(tblsizes, (d + a * hash_tbl_sizes)) / instances + np.multiply(b, tblsizes) + c
#             return np.multiply(cor_tbl_sizes, a) / instances + np.multiply(tblsizes, b) + c

    def fit_func_linear(x, a, b, c):
        tblsizes = [i[0] for i in x]
        instances = [i[1] for i in x]
        return np.multiply(tblsizes, a) / instances + np.multiply(tblsizes, b) + c

    if quadratic:
        fit_func = fit_func_quadratic
    else:
        fit_func = fit_func_linear

    data = fit_data.groupby(["rows", "instances"])["duration"].agg({"mean": np.mean, "var": np.var}).reset_index()
    data["rows100k"] = data["rows"] / 100000
    x = np.array(data[["rows100k", "instances"]])
    y = np.array(data["mean"])
    err = np.array(data["var"])
    
    if quadratic:
        output = curve_fit(fit_func_quadratic, x, y, sigma=err, full_output=True)
    else:
        output = curve_fit(fit_func, x, y, sigma=err, full_output=True)
    params = output[0]

    # Berechnung R^2
    infodict = output[2]
    ssErr = (infodict['fvec']**2).sum()
    ssTot = ((y-y.mean())**2).sum()
    rsquared = 1-(ssErr/ssTot )
    print "rsquared = %.3f" % rsquared
    
    # Print Chi-Square
    f_obs = y
    f_exp = fit_func(x, *params)
    chisq = np.sum(np.square(f_obs-f_exp)/err)
    dof = len(f_obs) - len(params)
    red_chisq = chisq/dof
    print "red_chisq = %.3f" % red_chisq
    
    # Print perr
    perr = np.sqrt(np.diag(output[1]))
    for i, single_perr in enumerate(perr):
        print "perr[%d] = %.3f" % (i, single_perr)
    

    
    plot_data["rows100k"] = plot_data["rows"] / 100000
    def do_plot():
        # Plot measured data
        plot_data.groupby(["rows100k", "instances"])['duration'].mean().unstack(level="rows100k").plot()

        # Plot fitted data
        for tbl_size, group in data.groupby("rows100k"):
            x_model = [(tbl_size, i) for i in range(1, 1025, 10)]
            y_model = fit_func(x_model, *params)
            plt.plot([x_val[1] for x_val in x_model], y_model, "--")
        
        if quadratic:
            # Extra plot
            def extra_func(x, a, b, c, d):
                tblsizes = np.array([i[0] for i in x])
                hash_tbl_sizes = np.array([i[1] for i in x])
                instances = np.array([i[2] for i in x])

                return np.multiply(tblsizes, (d + a * hash_tbl_sizes)) / instances + np.multiply(b, tblsizes) + c

            tbl_size = 300
            hash_tbl_size = 0.06
            x_model = [(tbl_size, hash_tbl_size, i) for i in range(1, 1025, 10)]
            y_model = extra_func(x_model, *params)
            plt.plot([x_val[2] for x_val in x_model], y_model, "--")



    plt.figure()
    do_plot()
    plt.yscale("log")

    plt.figure()
    do_plot()
    plt.xscale("log")
    plt.yscale("log")
    
    # Plot residuals for biggest row size
    biggest_row_size = data['rows'].max()
    crit = data['rows'] == biggest_row_size
    big_data = data[crit]
    big_x = np.array(big_data[["rows100k", "instances"]])
    big_y = np.array(big_data["mean"])
    exp_big_y = fit_func(big_x, *params)
    big_residuals = big_y - exp_big_y
    plt.figure()
    plt.plot([i[1] for i in big_x], big_residuals, "x")
    
    return params
```

```python
def fit_func_linear(x, a, b, c):
        tblsizes = [i[0] for i in x]
        instances = [i[1] for i in x]
        return np.multiply(tblsizes, a) / instances + np.multiply(tblsizes, b) + c

scan_params = [  3.85258544e-01,  -7.00891666e-05,   2.91533506e-01]
```

```python
x = [1] + range(2, 1025, 2)
full_x = [(300, i) for i in x]
y = fit_func_linear(full_x, *scan_params)
predicted = pandas.Series(y, index=x)
predicted.name = "predicted"

ref_scans = ref_df_scan[ref_df_scan["op_name"] == "TableScan"]
measured = ref_scans.groupby("instances")['duration'].mean()
measured.name = "measured"

plt.figure()
predicted.plot()
measured.plot(style="x")
plt.xscale("log")
plt.yscale("log")

# Prepare Data Frame
# instances TableScan_measured TableScan_predicted
# might need set datafile missing "NaN" for gnuplot
filename = "scan_fitting.csv"
pandas.concat([measured, predicted], axis=1).to_csv(filename, sep='\t', na_rep='NaN')
```
