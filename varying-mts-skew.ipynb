```python
def boxplot_oltp(df, op="ResponseTask", column='end'):
    crit = df['op_name']==op
    resp = df[crit]

    olap_olxp_queries = ["vldb_q10", "vldb_q11", "vldb_q12", "vldb_xselling"]
    crit = resp["query_name"].apply(lambda val: val not in olap_olxp_queries)
    resp = resp[crit]

    resp['escapedQueryName'] = resp['query_name'].apply(lambda val: val.replace("_", "\\_"))
    resp.boxplot(column=column, by="escapedQueryName")
```

```python
def boxplot_olap(df, op="ResponseTask", column='end'):
    crit = df['op_name']==op
    resp = df[crit]

    olap_queries = ["vldb_q10", "vldb_q11", "vldb_q12"]
    crit = resp["query_name"].apply(lambda val: val in olap_queries)
    resp = resp[crit]

    resp['escapedQueryName'] = resp['query_name'].apply(lambda val: val.replace("_", "\\_"))
    resp.boxplot(column=column, by="escapedQueryName")
```

```python
df_10ms.columns
```

```python
plt.figure()
boxplot_oltp(df_10ms)
plt.title("MTS==10ms")
plt.yscale("log")
plt.ylabel("ResponseTask endTime")

plt.figure()
boxplot_oltp(df_30ms)
plt.title("MTS==30ms")
plt.ylabel("ResponseTask endTime")
plt.yscale("log")
```

```python
def true_boxplot_oltp(df):
    crit = df["op_name"] == "RequestParseTask"
    # Since each ResponseTask follows its RequestParseTask without any interleaving, we will not group anything and save some time
    true_times = df[crit].reset_index(drop=True)
    crit2 = df["op_name"] == "ResponseTask"
    end_times = df[crit2]["start"].reset_index(drop=True)
    assert(len(true_times) == len(end_times))
    true_times['trueEnd'] = end_times - true_times["end"]
    true_times["respEnd"] = end_times

    olap_olxp_queries = ["vldb_q10", "vldb_q11", "vldb_q12", "vldb_xselling"]
    crit = true_times["query_name"].apply(lambda val: val not in olap_olxp_queries)
    true_times = true_times[crit]

    true_times['escapedQueryName'] = true_times['query_name'].apply(lambda val: val.replace("_", "\\_"))
    true_times.boxplot(column="trueEnd", by="escapedQueryName")
```

Plot of mean startTime of OLTP queries:

```python
resptasks = df[df["op_name"] == "ResponseTask"]
```

```python
resptasks['mts'] = resptasks['run'].apply(lambda row: int(row.split("_")[0]))
```

```python
olap_olxp = ['vldb_q10', 'vldb_q11', 'vldb_q12', 'vldb_xselling']
```

```python
crit = resptasks['query_name'].apply(lambda val: val in olap_olxp)
oltp_resp = resptasks[~crit]
```

```python
plt.figure()
oltp_resp.groupby('mts')['start'].mean().plot()
plt.title("OLTP Responsetime")
plt.ylabel("ReponseTask start time")
```
