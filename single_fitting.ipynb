```python
def get_weight(data, weigh_tbl_size=False):
    data.name = 'meanDur'
    tbl_size_100k = data.reset_index()['rows'].max() / 100000
    new_data = data.reset_index(level='rows', drop=True).reset_index()
    x = np.array(new_data['instances'])
    weights = np.concatenate([[0.001], np.diff(x)])
    weights = np.power(np.subtract(weights.max()+1, weights), 2)
#     weights = np.subtract(weights.max()+1, weights)
    if weigh_tbl_size:
        new_data['weights'] = weights * tbl_size_100k
    else:
        new_data['weights'] = weights
    return new_data
```

```python
def fit_func(x_tuples, a, b, c):
    instances = [x[0] for x in x_tuples]
    tablesizes = [x[1] for x in x_tuples]
    return np.multiply(a, tablesizes) / instances + np.multiply(b, tablesizes) + c
#     return np.multiply(a, tablesizes) / instances + b
```

```python
def fit_func_quadratic(x_tuples, a, b):
    instances = [x[0] for x in x_tuples]
    tablesizes = [x[1] for x in x_tuples]
    return np.multiply(a, tablesizes*(np.multiply(0.443,tablesizes))) / instances + np.multiply(b, tablesizes*(np.multiply(0.443, tablesizes))) 
```

```python
def plot_fitting(df, weigh_tbl_size=False, quadratic_fitting=False):
    meantimes = df.groupby(["rows", "instances"])["duration"].mean()
    fitdata = meantimes.groupby(level=0).apply(lambda x: get_weight(x, weigh_tbl_size=weigh_tbl_size)).reset_index()
    fitdata["rows_100k"] = fitdata["rows"]/100000
    x = np.array(fitdata[["instances", "rows_100k"]])
    y = np.array(fitdata["meanDur"])
    weights = fitdata["weights"]
    if quadratic_fitting:
        params, cov = curve_fit(fit_func_quadratic, x, y, sigma=weights)
    else:
        params, cov = curve_fit(fit_func, x, y, sigma=weights)
    y_model = fit_func(x, *params)
    from scipy.stats import chisquare
    red_chisq = chisquare(y, f_exp=y_model, ddof=len(params))[0]/(y.size-1-len(params))
    print "Red. chi-sq: %.3f" % red_chisq
    import matplotlib.pyplot as plt
    plt.figure(figsize=(20,20))
    plt.yscale("log")
    plt.ylabel("Mean Task Size in ms")
    for rows_100k, group in fitdata.groupby("rows_100k"):
        x_measure = np.arange(1, 1024,step=10)
        x_pred = [(i, rows_100k) for i in x_measure]
        plt.plot(x_measure, fit_func(x_pred, *params), "--", label="Fitted %d" % rows_100k)
        group.plot(x="instances", y="meanDur", label="Measured %d" % rows_100k)   
        plt.legend(loc="best")
```

```python
def fit_func(x, a, b, c):
    tblsizes = [d[0] for d in x]
    instances = [d[1] for d in x]
    return np.divide(np.multiply(np.square(tblsizes), a), instances) + np.multiply(b, tblsizes)

plt.figure()
plt.yscale("log")
weighted = times.groupby(level=0).apply(lambda x: get_weight(x))
params, cov = curve_fit(fit_func, np.array(weighted.index.get_values()), np.array(times))
for tbl_size, group in times.groupby(level="rows"):
    y_model = fit_func(np.array(group.index.get_values()), *params)    
    local_index_group = group.reset_index(level='rows', drop=True)
    local_index_group.plot()
    plt.plot(local_index_group.index, y_model, "--")
```
